# Stack implementation with classes from scratch

```python
class stack:
    def __init__(self):
        self.data = []
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.size
    
    def push(self,value):
        self.data.append(value)
        self.size+=1
        
    def top(self):
        return self.data[-1]
    
    def pop(self):
        r = self.data[-1]
        del self.data[-1]
        self.size-=1
        return r
    
    def display(self):
        return self.data
```

```python
s = stack()
print(type(s))
```

```python
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.display()
```

```python
s.pop()
s.display()
```

```python
s.top()
```

# Srack program in user view

```python
class stack:
    def __init__(self):
        self.data = []
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.size
    
    def push(self,value):
        self.data.append(value)
        self.size+=1
        
    def top(self):
        if self.size==0:
            print("Stack is empty")
        else:
            print(self.data[-1])
    
    def pop(self):
        if self.size==0:
            print("Stack is empty")
        else:
            r = self.data[-1]
            del self.data[-1]
            self.size-=1
            print(r,"deleted")
        
    def display(self):
        if self.size==0:
            print("Stack is empty")
        else:
            print(self.data)
        
 
s=stack()   
a=0
while a!=5:
        
        print("\n\n-----Stack------\nPress 1 for insert element in stack\nPress 2 for delete a element from the stack\nPess 3 for display\nPress 4 for top element of the stack\nPress 5 for break")
        a=int(input("Press a key to perform opreation:"))
        if a==1:
            i=int(input("Enter a number you want to insert:"))
            s.push(i)
            print(i,"inserted.")
       
        elif a==2:
            s.pop()
            
        
        elif a==3:
            s.display()
        
        elif a==4:
            s.top()
        
        elif a==5:
            pass
        
        else:
            print("Press a valid Key!")
                    
```
