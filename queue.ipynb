## Queue implementation with scratch.

```python
class Queue:
    
    def __init__(self):
        self.data = []
        self.size = 0
        self.front = -1
        self.rare = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.size
    
    def enqueue(self,element):
        self.data.append(element)
        self.front+=1
        self.size+=1
    
    def dequeue(self):
        if self.size ==0:
            print("Queue is empty.")
        else:
            r=self.data[self.rare]
            del self.data[self.rare]
            print(r,"deleted")
        
    def display(self):
        if self.size==0:
            print("Queue is empty.")
        else:
            print(self.data)
    
    def rear(self):
        if self.size==0:
            print("Queue is empty.")
        else:
            print("First element in the queue is",self.data[self.rare])
    
    def font(self):
        if self.size==0:
            print("Queue is empty.")
        else:
            print("Last element in the queue is",self.data[self.front])
            
q=Queue()

q.display()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
q.font()
q.rear()
q.display()
q.dequeue()
q.display()
q.rear()
```
